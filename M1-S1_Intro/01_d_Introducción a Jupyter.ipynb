# Introducción a Colab

Colaboratory es un entorno gratuito de blocs de notas de Jupyter que no requiere configuración y se ejecuta completamente en la nube. 

Colaboratory te permite ejecutar código python en el navegador con un solo clic.
desde cualquier cuenta de google en la dirección: (https://colab.research.google.com/)

El Notebook también se ha ejecutado en un entorno de con instalación en anaconda y ejecutado mediante el editor de Visual Studio Code con el plugin python

Desde las celdas de código se puede ejecutar lineas de ejecución de CMD (windows) o de terminal (linux).  
Para esto se utiliza el carácter *!* seguido de la sentencia.  
Es una buena práctica referenciar la versión de las librerias por ejemplo para conocer la versión de Python:  

```bash
! python --version
```

````
! python --version
````

En caso de ejcutar en un entorno local de anaconda (Jupyter Notebook o VS code)
```bash
! conda --version
```

````
! conda --version
````

## Comandos Magicos 

Son comandos que se ejecutan sobre el kernel de IPython y agregan ciertas funcionalidades adicionales al código. Por ejemplo mostrar el tiempo de ejecución.   
Los comandos utilizan prefijo % cuando se ejecutan en una sola linea, o %% para toda la celda.
Para conocer todos los comandos disponibles:
```python
%lsmagic
```

En el siguiente artículo se presentan algunos comandos útiles:
https://towardsdatascience.com/top-10-magic-commands-in-python-to-boost-your-productivity-1acac061c7a9


````
%lsmagic
````

Algunos de los comandos más usados son:
```python
%matplotlib inline
%%time
%env
```


## Atajos de Teclado

| Acción                      | COLAB            | Jupyter |
|-----------------------------|------------------|---------|
| Mostrar atajos de teclado   | Ctrl/Cmd + M + H | H       |
| Insertar celda arriba       | Ctrl/Cmd + M + A | A       |
| Insertar celda abajo        | Ctrl/Cmd + M + B | B       |
| Eliminar celda seleccionada | Ctrl/Cmd + M + D | DD      |
| Interrumpir ejecución       | Ctrl/Cmd + M + I | II      |
| Convertir a código          | Ctrl/Cmd + M + Y | Y       |
| Convertir a Texto           | Ctrl/Cmd + M + M | M       |
